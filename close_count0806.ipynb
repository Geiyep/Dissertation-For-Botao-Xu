```python
import pandas as pd
import numpy as np
import sys
```

```python
data = pd.read_excel("close.xlsx")
data.set_index("Date",inplace=True)
```

```python
company = [i for i in data.columns]
col_name = ["Date"]+company+["total"]
result = pd.DataFrame(columns=col_name)
```

```python
care = [i for i in data.index if i.month == 6 or i.month==12]

for i in range(2010,2023):
    date_check = [j for j in care if j.year==i]    
    if not i == 2010:
        close_dict = {}
        count = 0
        if pd.Timestamp(str(i)+"-06-30") in date_check:
            date = pd.Timestamp(str(i)+"-06-30")
        elif pd.Timestamp(str(i)+"-06-29") in date_check:
            date = pd.Timestamp(str(i)+"-06-29")
        elif pd.Timestamp(str(i)+"-06-28") in date_check:
            date = pd.Timestamp(str(i)+"-06-28")
        elif pd.Timestamp(str(i)+"-06-27") in date_check:
            date = pd.Timestamp(str(i)+"-06-27")
        elif pd.Timestamp(str(i)+"-06-26") in date_check:
            date = pd.Timestamp(str(i)+"-06-26")
        else:
            print("Near "+str(i)+"-06-30, all three days not in index")
            sys.exit()
        close_dict["Date"] = date
        for c in company:
            if not pd.isnull(data.loc[date][c]):
                close_dict[c] = 1
                count += 1
            else:
                close_dict[c] = 0
        close_dict["total"] = count
        result = result.append(close_dict,ignore_index=True)

    if not i == 2022:
        close_dict = {}
        count = 0
        if pd.Timestamp(str(i)+"-12-31") in date_check:
            date = pd.Timestamp(str(i)+"-12-31")
        elif pd.Timestamp(str(i)+"-12-30") in date_check:
            date = pd.Timestamp(str(i)+"-12-30")
        elif pd.Timestamp(str(i)+"-12-29") in date_check:
            date = pd.Timestamp(str(i)+"-12-29")
        elif pd.Timestamp(str(i)+"-12-28") in date_check:
            date = pd.Timestamp(str(i)+"-12-28")
        elif pd.Timestamp(str(i)+"-12-27") in date_check:
            date = pd.Timestamp(str(i)+"-12-27")
        else:
            print("Near "+str(i)+"-12-31, all three days not in index")
            sys.exit()
        close_dict["Date"] = date
        for c in company:
            if not pd.isnull(data.loc[date][c]):
                close_dict[c] = 1
                count += 1
            else:
                close_dict[c] = 0
        close_dict["total"] = count
        result = result.append(close_dict,ignore_index=True)

result
```
