```python
import pandas as pd
import numpy as np
import sys
```

```python
net_prof = pd.read_excel("0815/data/0815/netprofitmagin.xlsx")
gross_prof = pd.read_excel("0815/data/0815/grossprofitmagin.xlsx")
roe = pd.read_excel("roe4season.xlsx")

net_prof.set_index("Date",inplace=True)
gross_prof.set_index("Date",inplace=True)
roe.set_index("Date",inplace=True)
```

```python
# gross_prof>20, net_prof>10, roe>10
company = [i for i in net_prof.columns]
dates = [i for i in net_prof.index]

results = {}
for date in dates:
    result = []
    for c in company:
        if not pd.isnull(net_prof.loc[date][c]) and not pd.isnull(gross_prof.loc[date][c]) and not pd.isnull(roe.loc[date][c]):
            if net_prof.loc[date][c] > 10 and roe.loc[date][c]>10 and gross_prof.loc[date][c]> 20:
                result = result + [c]
    results[date] = result
    print(date)
    print(len(result))
    print(result)
```
