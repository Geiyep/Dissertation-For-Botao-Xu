```python
from datetime import datetime
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
import sys
```

```python
df_closes = pd.read_excel("closes.xlsx")
df_high = pd.read_excel("high.xlsx")
df_low = pd.read_excel("low.xlsx")
df_spa = pd.read_excel("SPYA.xlsx")

df_closes.set_index("Date",inplace=True)

df_spa.set_index("Date",inplace=True)
```

```python
def ma_helper(closes):
    return np.sum(closes)/len(closes)

def momentum(closes):
    returns = np.log(closes)
    x = np.arange(len(returns))
    slope, _, rvalue, _, _ = linregress(x,returns)
    return ((1+slope)**252)*(rvalue**2)
```

```python
#SPY从200天前开始，close从100天前开始，atr20从100-20=80天前开始，momentum从100-90=10天前开始，都有offset
company = [i for i in df_closes.columns]
momentums = df_closes.copy(deep=True)
for c in company:
    momentums[c] = df_closes[c].rolling(90).apply(momentum, raw=False)

momentums.to_excel("momentums.xlsx")
```

```python

sma100 = df_closes.copy(deep=True)
for c in company:
    sma100[c] = df_closes[c].rolling(100).apply(ma_helper, raw=False)

sma100.to_excel("sma100.xlsx")
```

```python
sma200 = df_spa.copy(deep=True)
sma200['close'] = df_spa['close'].rolling(200).apply(ma_helper, raw=False)

sma200.to_excel("sma200.xlsx")
```

```python
df_closes = pd.read_excel("closes.xlsx")
df_high = pd.read_excel("high.xlsx")
df_low = pd.read_excel("low.xlsx")
```

```python
df_max_value = df_closes.copy(deep=True)
for i in range(1,len(df_closes)):
    for c in company:
        high = df_high.loc[i][c]
        low = df_low.loc[i][c]
        cp = df_closes.loc[i-1][c]
        v1 = high-low
        vc = np.array([high-low, np.abs(high-cp),np.abs(low-cp)])
        vm = np.max(vc)
        df_max_value.loc[i,c] = vm
df_max_value
```

```python
atr20 = df_max_value.copy(deep=True)
for c in company:
    atr20[c] = df_max_value[c].rolling(20).apply(ma_helper, raw=False)

atr20.to_excel("atr20.xlsx")
```
